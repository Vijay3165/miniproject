# Credit Card Fraud Detection - Logistic Regression

Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
#from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import StandardScaler
```

This Dataset is highly imbalanced

0 --> Normal Transaction

1 --> fraudulent transaction

```python
# Data Cleaning: Removing missing values
data = credit_card_data.dropna()
print(data)
```

```python
# Data Formatting: Ensuring column names are consistent
data.columns = data.columns.str.strip().str.lower()
```

```python
# Splitting features and target variable from the dataset 
X = data.drop(columns=['class'])  # Assuming 'class' is the target variable
y = data['class']
print(X.shape[1])
```

```python
print(y)
```

```python

# Define SMOTE model
smote = SMOTE(sampling_strategy=1, random_state=42)  # 1.0 means equal fraud and non-fraud cases
# Apply SMOTE to generate synthetic fraud transactions
X_resampled, y_resampled = smote.fit_resample(X, y)
# Check the new class distribution
print(pd.Series(y_resampled).value_counts())
```

```python
# Feature Selection: Selecting top 10 features
selector = SelectKBest(score_func=f_classif, k=10)
X_selected = selector.fit_transform(X_resampled, y_resampled)
#X_selected = selector.fit_transform(X, y)
print(X_selected.shape)
mask = selector.get_support()

# Get the feature names
selected_features = X.columns[mask]

print("Selected Features:")
print(selected_features.tolist())
```

```python
# Feature Scaling: Standardizing numerical values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([X_resampled, y_resampled], axis=1)

#new_dataset = pd.concat([X, y], axis=1)
```

Splitting the data into Features & Targets

```python
# Splitting features and target correctly
X = new_dataset.drop(columns='class', axis=1)
Y = new_dataset['class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
# Now splitting into training & test sets
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression(class_weight={0: 1, 1: 3}, solver='saga', max_iter=20000)
```

```python
X_train_sample, _, Y_train_sample, _ = train_test_split(X_train, Y_train, random_state=42)
model.fit(X_train_sample, Y_train_sample)
```

Model Evaluation

Evaluating the model

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
training_data_precision = precision_score(Y_train,X_train_prediction)
training_data_recall = recall_score(Y_train,X_train_prediction)
training_data_f1 = f1_score(Y_train,X_train_prediction)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
test_data_precision = precision_score(Y_test,X_test_prediction)
test_data_recall = recall_score(Y_test,X_test_prediction)
test_data_f1 = f1_score(Y_test,X_test_prediction)
```

```python


# Predict on test data
y_pred = model.predict(X_test)

# Generate confusion matrix
cm = confusion_matrix(Y_test, y_pred)

# Extract TN, FP, FN, TP from confusion matrix
TN, FP, FN, TP = cm.ravel()

# Print the confusion matrix values
print("Confusion Matrix:")
print(cm)
print(f"\nTrue Positives (TP): {TP}")
print(f"True Negatives (TN): {TN}")
print(f"False Positives (FP): {FP}")
print(f"False Negatives (FN): {FN}")

# Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap=plt.cm.Blues)

plt.title("Confusion Matrix - Test Dataset")
plt.show()
```

```python
print(f'Accuracy score on Training Data : {training_data_accuracy:.2%}')
print(f'Precision score on Training data : {training_data_precision:.2%}')
print(f'Recall score on Training data : {training_data_recall:.2%}')
print(f'f1 score on Training data : {training_data_f1:.2%}')
```

```python
print(f'Accuracy score on Test Data : {test_data_accuracy:.2%}')
print(f'Precision score on Test data : {test_data_precision:.2%}')
print(f'Recall score on Test data : {test_data_recall:.2%}')
print(f'f1 score on Test data : {test_data_f1:.2%}')
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Predict probabilities
y_prob = model.predict_proba(X_test)[:, 1]
threshold = 0.8
y_pred_custom = (y_prob > threshold).astype(int)
# Compute new AUC
auc_score = roc_auc_score(Y_test, y_pred_custom)
print(f"ROC AUC Score: {auc_score}")
# Plot ROC Curve
fpr, tpr, _ = roc_curve(Y_test, y_pred_custom)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {auc_score:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
#plt.axhline(y=threshold, color='navy', linestyle='--', label=f"Threshold {threshold}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Get prediction probabilities for the test set
y_test_probs = model.predict(X_test).ravel() 
# Compute precision-recall pairs
precision, recall, thresholds = precision_recall_curve(Y_test, y_test_probs)
# Compute the average precision score (area under the precision-recall curve)
avg_precision = average_precision_score(Y_test, y_test_probs)

# Plot the precision-recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label=f'PR Curve (AP = {avg_precision:.4f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid(True)
plt.show()
```
